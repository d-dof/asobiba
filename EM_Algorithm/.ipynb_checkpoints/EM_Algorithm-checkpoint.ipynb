```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
from scipy.stats import norm
from numpy import linalg as LA

data_number_1 = 500
data_number_2 = 300
data_number_3 = 400

data_number = data_number_1 + data_number_2
toy_data = []
for i in range(data_number_1):
    toy_data.append(normal(10, 10))
for i in range(data_number_2):
    toy_data.append(normal(50, 20))
for i in range(data_number_3):
    toy_data.append(normal(100, 10))


toy_data = np.array(toy_data)

plt.subplot(2, 1, 1)
plt.hist(toy_data, bins=50)
#plt.show()

def normal_pdf(x, mu, sigma):
    return (1.0 / np.sqrt(2 * np.pi * (sigma))) * np.exp(- (x - mu)**2 / sigma)

class Theta():
    def __init__(self, cluster_number):
        self.pi = np.array([1.0 / cluster_number]*cluster_number)
        self.mu = np.array(np.random.rand(cluster_number) * 10.0)
        self.sigma = np.array([10.0]*cluster_number)
        self.cluster_number = cluster_number
        
def gamma(x, theta):
    n = len(x)
    k = theta.cluster_number    
    G = np.empty((n, k))
    for t in range(n):
        for i in range(k):
            a = theta.pi[i] * normal_pdf(x[t], theta.mu[i], theta.sigma[i])
            if a < 1.0e-20:  # 小さすぎる値を調整
                a = 0.0
            G[t][i] = a
        if np.sum(G[t]) == 0.0:  # ある系列が何も起こらないなら確率を等しくしてしまえば良い
            G[t] = np.array([1.0 / k] * k)
        else:
            G[t] = G[t] / np.sum(G[t])
    return G

def em(x, cluster_number, eps):
    delta = float("inf")
    n = len(x)
    j = 0
    theta_old = Theta(cluster_number)
    while delta > eps:
        theta_new = Theta(cluster_number)
        G = gamma(x, theta_old)
        G = G.T
        for i in range(cluster_number):
            a = np.sum(G[i])
            b = np.sum(G[i] * x)
            theta_new.pi[i] = (1.0 / n) * a
            theta_new.mu[i]= (1.0 / a) * b
            theta_new.sigma[i]= (1.0 / a) * np.sum(G[i] * (x - theta_new.mu[i]) ** 2)
        delta = LA.norm(theta_new.mu - theta_old.mu)
        theta_old = theta_new
        j += 1
    print("回った回数は", j, "回")
    return theta_new

x = toy_data
cluster_number = 3
a = em(x, cluster_number, 0.005)
for i in range(a.cluster_number):
    print("重み付けは", a.pi[i], "で N(", a.mu[i], ",", a.sigma[i], ")と推定")

x = range(int(toy_data.min()), int(toy_data.max()))
plt.subplot(2, 1, 2)
for i in range(a.cluster_number):
        y = [a.pi[i] * normal_pdf(x_j, a.mu[i], a.sigma[i]) for x_j in x]
        plt.plot(x, y)
plt.show()
```

```python
def aic(x, theta):
    aic = 0.0
    cluster_number = theta.cluster_number
    model_complexity = 3 * cluster_number
    aic = model_complexity
    mu = theta.mu
    sigma = theta.sigma
    pi = theta.pi
    for x_t in x:
        aic -= np.log(np.sum([pi[i] * normal_pdf(x_t, mu[i], sigma[i]) for i in range(cluster_number)]))
    return aic

def bic(x, theta):
    bic = 0.0
    cluster_number = theta.cluster_number
    model_complexity = 3 * cluster_number * np.log(len(x))
    bic = model_complexity
    mu = theta.mu
    sigma = theta.sigma
    pi = theta.pi
    for x_t in x:
        bic -= np.log(np.sum([pi[i] * normal_pdf(x_t, mu[i], sigma[i]) for i in range(cluster_number)]))
    return bic
```

```python
def model_selection(x, max_cluster_number=10):
    aic_old = float("inf")
    theta_old = Theta(1)
    eps = 0.005
    for i in range(1, max_cluster_number + 1):
        print(i)
        theta_new = em(x, i, eps)
        for j in range(theta_new.cluster_number):
            print("重み付けは", theta_new.pi[j], "で N(", theta_new.mu[j], ",", theta_new.sigma[j], ")と推定")
        aic_new = bic(x, theta_new)
        print("AICは", aic_new)
        xx = range(int(x.min()), int(x.max()))
        for i in range(theta_new.cluster_number):
            y = [theta_new.pi[i] * normal_pdf(x_j, theta_new.mu[i], theta_new.sigma[i]) for x_j in xx]
            plt.plot(xx, y)
        plt.show()
        if aic_new > aic_old:
            return theta_old
        theta_old = theta_new
        aic_old = aic_new 
    return theta_old
```

```python
theta = model_selection(toy_data, 5)
```
